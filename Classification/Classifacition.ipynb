```python
y = df["Type"].values
x = df.drop("Type",axis = 1).values 
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

## Shift + Tab dökümantasyonu açar.

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

#Column sayısını verdik ilk Dense' mize 
#Çıkış kolonuna 1 verip aralarında ki değerlere de ikisinin arasında bir değer veriyoruz (30 ile 1 arası)
model.add(Dense(units = 30, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
#0 ile 1 değer verdiği için sigmoid kullandık (classifacition)
model.add(Dense(units = 1, activation = "sigmoid"))
#binary_crossentropy = ikili değer vermesini istiyoruz 
model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
model.fit(x = x_train, y = y_train, epochs = 700,validation_data =(x_test,y_test), verbose = 1)
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
```

```python
#Epoch fazla verdiğimizden ötürü overfitting yapıyor.
kayipVerisi.plot()
```

```python
model = Sequential()

#Column sayısını verdik ilk Dense' mize 
#Çıkış kolonuna 1 verip aralarında ki değerlere de ikisinin arasında bir değer veriyoruz (30 ile 1 arası)
model.add(Dense(units = 30, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
model.add(Dense(units = 15, activation = "relu"))
#0 ile 1 değer verdiği için sigmoid kullandık (classifacition)
model.add(Dense(units = 1, activation = "sigmoid"))
#binary_crossentropy = ikili değer vermesini istiyoruz 
model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

## earlyStopping = erken durdurma epoch kafamıza göre veriyoruz
## callbacks = earlyStopping yaparsak en iyi şekilde durduruyor 

```python
earlyStopping = EarlyStopping(monitor="val_loss", mode="min",verbose=1,patience = 25)
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

## Dropout

```python
model = Sequential()

#Column sayısını verdik ilk Dense' mize 
#Çıkış kolonuna 1 verip aralarında ki değerlere de ikisinin arasında bir değer veriyoruz (30 ile 1 arası)
model.add(Dense(units = 30, activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units = 15, activation = "relu"))
model.add(Dropout(0.6))

model.add(Dense(units = 15, activation = "relu"))
model.add(Dropout(0.6))

#0 ile 1 değer verdiği için sigmoid kullandık (classifacition)
model.add(Dense(units = 1, activation = "sigmoid"))
#binary_crossentropy = ikili değer vermesini istiyoruz 
model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
modelKayip = pd.DataFrame(model.history.history)
```

```python
modelKayip.plot()
```

```python
tahminlerimiz = model.predict_classes(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,tahminlerimiz))
```

```python
print(classification_report(y_test,tahminlerimiz))
```
